```python
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder(
    categories=[[
        "Month-to-month",
        "One year",
        "Two year"
    ]]
)

data[["Contract"]] = oe.fit_transform(
    data[["Contract"]]
)
```

```python
#remening all cat. are nominal cat. used-->onehotencoder
```

```python
data.columns
```

```python
nominal_cat_cols = [
    'gender','Partner','Dependents',
    'PhoneService','MultipleLines',
    'InternetService','OnlineSecurity',
    'OnlineBackup','DeviceProtection',
    'TechSupport','StreamingTV',
    'StreamingMovies','PaperlessBilling',
    'PaymentMethod'
]
```

```python
from sklearn.preprocessing import OneHotEncoder

# Create OneHotEncoder object
ohe = OneHotEncoder(
    drop='first',
    handle_unknown='ignore',
    sparse_output=False
)
```

```python
nominal_cols_transformed= ohe.fit_transform(data[nominal_cat_cols])
```

```python
# nominal_cols_transformed
"""
Output is a NumPy array because sklearn's OneHotEncoder returns a NumPy array after fit_transform(). 
Hence, it was converted into a DataFrame for easier concatenation and column handling."""
```

```python
nominal_cols_df=pd.DataFrame(nominal_cols_transformed,columns=ohe.get_feature_names_out(),index=data.index)
```

```python
nominal_cols_df.head()
```

```python
data.head(1)
```

```python
#Extract remening column

extract_rem_data=data[["customerID","SeniorCitizen","tenure","Contract","MonthlyCharges","TotalCharges","Churn"]]
```

```python
extract_rem_data.head(2)
```

```python
#concat both rem + nominal cat data transformer
```

```python
final_trf=pd.concat([extract_rem_data,nominal_cols_df],axis=1)
```
